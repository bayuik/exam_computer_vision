```python
from PIL import Image
image_file = r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro.jpg"

def get_pixel_size(image):
    width, height = Image.open(image).size
    return width * height

pixel_size = get_pixel_size(image_file)
print('1. Ukuran Pixel adalah :', pixel_size)
```

```python
def get_center_coordinate(image):
    width, height = Image.open(image).size
    return (width/2, height/2)

def get_rgb_composition(image):
    width, height = Image.open(image).size
    rgb_image = Image.open(image).convert('RGB')
    red = 0
    green = 0
    blue = 0
    for i in range(width):
        for j in range(height):
            r, g, b = rgb_image.getpixel((i, j))
            red += r
            green += g
            blue += b
    return (red, green, blue)

rgb_composition = get_rgb_composition(image_file)
center_coordinate = get_center_coordinate(image_file)

print('2. Koordinat titik pusat adalah :', center_coordinate)
print('2. Komposisi RGB adalah :', rgb_composition)
print('Red :', rgb_composition[0])
print('Green :', rgb_composition[1])
print('Blue :', rgb_composition[2])
```

```python
def convert_to_grayscale(image):
    image = Image.open(image)
    width, height = image.size
    new_image = Image.new('L', (width, height))
    for x in range(width):
        for y in range(height):
            r, g, b = image.getpixel((x, y))
            gray = (r + g + b) // 3
            new_image.putpixel((x, y), gray)
    return new_image

grayscale_image = convert_to_grayscale(image_file)
grayscale_image.save(r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro_grayscale.jpg")
print("3. Covert image to grayscale")
grayscale_image
```

```python
grayscale_image_file = r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro_grayscale.jpg"

def convert_to_bw(image):
    img = Image.open(image)
    bw_image = img.point(lambda x: 0 if x<100 else 1, '1')
    return bw_image

bw_image = convert_to_bw(grayscale_image_file)
bw_image.save(r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro_bw.jpg")
print('4. Convert image to black and white')
bw_image
```

```python
bw_image_file = r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro_bw.jpg"

def zoom_bw_image(image):
    img = Image.open(image)
    zoom_image = img.resize((img.width*4, img.height*4))
    return zoom_image

zoom_image = zoom_bw_image(bw_image_file)
zoom_image.save(r"C:\bayu_files\UTS\Bayu Indra Kusuma - pcd-fikonaro_zoom.jpg")
print('5. Zoom image')
zoom_image
```
